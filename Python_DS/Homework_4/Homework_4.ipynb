Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов

documents = [
{‘type’: ‘passport’, ‘number’: ‘2207 876234’, ‘name’: ‘Василий Гупкин’},
{‘type’: ‘invoice’, ‘number’: ‘11-2’, ‘name’: ‘Геннадий Покемонов’},
{‘type’: ‘insurance’, ‘number’: ‘10006’, ‘name’: ‘Аристарх Павлов’}
]

2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

directories = {
‘1’: [‘2207 876234’, ‘11-2’],
‘2’: [‘10006’],
‘3’: []
}

Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды “q”).

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def start_():    
    cmd = input('Введите команду: ')    
    while cmd.lower() != 'q':
        # Для as/ds/m нужен номер полки        
        if cmd.lower() == 'as' or cmd.lower() == 'ds' or cmd.lower() == 'm':           
            shelf = input('Введите номер полки: ') 
        #Для p/s/ad/d/m нужен номер документа
        if cmd.lower() == 'p' or cmd.lower() == 's' or cmd.lower() == 'ad' or cmd.lower() == 'd' or cmd.lower() == 'm':
            doc_num = input('Введите номер документа ')
        #Команды         
        if cmd.lower() == 'p':
            if find_doc(doc_num) is None:
                print('Документ не найден в базе')
            else:
                print(find_doc(doc_num))
        if cmd.lower() == 's':   
            if which_shelf(doc_num) is not None:
                print(f'Документ хранится на полке: {which_shelf(doc_num)}')
            else:
                print('Документ не найден в базе')
        if cmd.lower() == 'l':   
            all_info()
        if cmd.lower() == 'as':
            add_shelf(shelf)
        if cmd.lower() == 'ds':
            del_shelf(shelf)
        if cmd.lower() == 'ad':
            add_doc(doc_num)
        if cmd.lower() == 'd':
            doc_del(doc_num)
        if cmd.lower() == 'm':
            move_(doc_num, shelf)
        cmd = input('Введите команду: ')
```

```python
start_()
```

Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру
Примеры работы:

Введите команду:
p

Введите номер документа:

Результат:

Введите команду:

Введите номер документа:

Результат:

```python
def find_doc(doc_num):
    for doc in documents:
        if list(doc.items())[1][1] == doc_num:
            return list(doc.items())[2][1]
```

Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
Примеры работы:

Введите команду:
s

Введите номер документа:
10006

Результат:
Документ хранится на полке: 2

Введите команду:
p

Введите номер документа:
12345

Результат:
Документ не найден в базе

```python
def which_shelf(doc_num):
    for key, val in directories.items():
        for el in val:
            if doc_num == el:
                return key
```

Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам
Пример работы:

Введите команду:
l
Результат:

№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

```python
def all_info():
    for doc in documents:
        doc_num = doc['number']
        type_ = doc['type']
        name = doc['name']
        ws = which_shelf(doc_num)
        print(f'№: {doc_num}, тип: {type_}, владелец: {name}, полка хранения: {ws}')
```

Пункт 4. Пользователь по команде “as” может добавить новую полку
Примеры работы:

Введите команду:
as

Введите номер полки:
10

Результат:
Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.

Введите команду:
as

Введите номер полки:
1

Результат:
Такая полка уже существует. Текущий перечень полок: 1, 2, 3.

```python
# Перечень полок
def shelves():
    Print_txt = str(list(directories))
    return Print_txt.replace("'", '')[1:-1]

# Поиск полки
def exist_shelf(shelf):
    if shelf in list(directories):
        return True    
           
def add_shelf(shelf):
    if exist_shelf(shelf) == True:        
        print('Такая полка уже существует. Текущий перечень полок: ' + shelves() + '.')    
    else:
        directories[str(shelf)] = []
        print('Полка добавлена. Текущий перечень полок: ' + shelves() + '.')
```

Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
Примеры работы:

```python
def del_shelf(shelf):
    if len(directories[shelf]) == 0:
        del directories[shelf]
        print('Полка удалена. Текущий перечень полок: ' + shelves() + '.')
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3' + shelves() + '.')
```

```python
# Удалить документ из директории
def del_doc_in_dir(doc_num):
    if doc_num is None:
        print('Необходим номер документа')
    else:        
        del_var = which_shelf(doc_num)
        dir_upd = directories[del_var]
        dir_upd.remove(doc_num)    
        directories[del_var] = dir_upd

def doc_del(doc_num):
    i = 0
    for doc in documents:
        if doc['number'] == doc_num:
            is_del = 1
            del documents[i] 
            print('Документ удален. Текущий список документов:')
            all_info()
            del_doc_in_dir(doc_num)
            break
        i += 1
        is_del = 0
    if is_del == 0:
        print('Документ не найден в базе. Текущий список документов:')
        all_info()
```

```python
def move_(doc_num, shelf):
    check1 = find_doc(doc_num)
    check2 = exist_shelf(shelf)
    if check1 is not None and check2 is not None:
        if which_shelf(doc_num) == shelf:
            print(f'Документ уже находится на полке {shelf}, пермещение не требуется')
        else:
            del_doc_in_dir(doc_num)
            add_doc_in_dir(doc_num, shelf)
            print('Документ перемещен. Текущий список документов:')     
            all_info()
    else:
        if check1 is None:
            print('Документ не найден в базе. Текущий список документов:')
            all_info()
        if check2 is None:
            print('Такой полки не существует. Текущий перечень полок ' + shelves() + '.')  
```
