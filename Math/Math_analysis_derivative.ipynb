### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
from math import *
from scipy.misc import derivative
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: x**5 + 4*sin(2*x) + cos(3*x + 3)

print('Первая производная в точке х=1:',derivative(f, 1.0, dx=1e-6, n=1))
print('Вторая производная в точке х=1:',derivative(f, 1.0, dx=1e-6, n=2))
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
f = lambda x: (sin(2*x + 1))**5
x = np.linspace(-5, 5, num = 1000)

dx = []
ddx = []
for i in x:
  dx.append(derivative(f, i, dx=1e-6, n=1))
  ddx.append(derivative(f, i, dx=1e-6, n=2))

plt.plot(x, dx, label='Первая производная')
plt.plot(x, ddx, label='Вторая производная')
plt.grid()
plt.legend()
plt.xlabel('x')
plt.show()
```
